```python
import numpy as np
import pandas as pd
import datetime

# Generate dates for the past 3 months with a 2-day interval
start_date = datetime.date.today() - datetime.timedelta(days=3*30)
end_date = datetime.date.today()
dates = pd.date_range(start=start_date, end=end_date, freq='2D')

# Initialize a dictionary to store progress data for each finger
progress_data = {'Date': dates}

# Generate progress data for each finger
for finger in ['Thumb', 'Index', 'Middle', 'Ring', 'Pinky']:
    # Start value for the finger
    start_value = 7
    
    # Generate gradual increase over time with fluctuations
    trend = np.arange(len(dates)) * 0.1  # Gradual increase
    fluctuations = np.random.normal(loc=0, scale=5, size=len(dates))  # Random fluctuations
    
    # Generate progress data for the finger
    finger_data = start_value + trend + fluctuations
    
    # Ensure values are between 0 and 100
    finger_data = np.clip(finger_data, 0, 100)
    
    # Store progress data for the finger
    progress_data[finger] = finger_data

# Create a DataFrame from the progress data
progress_df = pd.DataFrame(progress_data)

# Display the DataFrame
print(progress_df)
```
